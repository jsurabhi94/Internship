### Q1.
Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
# opening the website
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
time.sleep(2)
```

```python
# creating empty lists
rank = []
name = []
artist = []
date = []
views = []
```

```python
# scraping data

try:
    r = driver.find_elements(By.XPATH, '//div[@class="mw-parser-output"]/table[2]/tbody/tr/td[1]')
    for i in r:
        rank.append(i.text)
except NoSuchElementException:
    rank.append('-')
    
try:
    n = driver.find_elements(By.XPATH, '//div[@class="mw-parser-output"]/table[2]/tbody/tr/td[2]')
    for i in n:
        name.append(i.text)
except NoSuchElementException:
    name.append('-')
    
try:
    ar = driver.find_elements(By.XPATH, '//div[@class="mw-parser-output"]/table[2]/tbody/tr/td[3]')
    for i in ar:
        artist.append(i.text)
except NoSuchElementException:
    artist.append('-')
    
try:
    d = driver.find_elements(By.XPATH, '//div[@class="mw-parser-output"]/table[2]/tbody/tr/td[5]')
    for i in d:
        date.append(i.text)
except NoSuchElementException:
    date.append('-')
    
try:
    v = driver.find_elements(By.XPATH, '//div[@class="mw-parser-output"]/table[2]/tbody/tr/td[4]')
    for i in v:
        views.append(i.text)
except NoSuchElementException:
    views.append('-')
```

```python
# printing the length of arrays

print(len(rank), len(name), len(artist), len(date), len(views))
```

```python
# creating dataframe
most_viewed = pd.DataFrame({'Rank':rank, 'Name':name, "Artist":artist, "Upload_date":date, "Views (in billions)":views})
most_viewed
```
