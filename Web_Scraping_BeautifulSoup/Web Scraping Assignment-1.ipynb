### Q1.  Write a python program to display all the header tags from wikipedia.org and make data frame.

```python
url = "https://en.wikipedia.org/wiki/Main_Page"
page = requests.get(url)
soup = BeautifulSoup(page.content)

headlines = []
for i in soup.find_all('span', class_='mw-headline'):
    headlines.append(i.text)
    
Head_df = pd.DataFrame(headlines, columns=['Headline'])
Head_df
```

### Q2. Write a python program to display IMDB’s Top rated 50 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
url = "https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&view=advanced"
page = requests.get(url)
soup = BeautifulSoup(page.content)

# creating empty lists
name = []
rating = []
year = []

for i in soup.find_all('h3', class_='lister-item-header'):
    name.append(i.text.strip().split('\n')[1])
for rate in soup.find_all('div', class_='inline-block ratings-imdb-rating'):
    rating.append(rate.text.strip())
for y in soup.find_all('h3', class_='lister-item-header'):
    year.append(y.text.strip().split('\n')[2])
```

```python
imdb_50 = pd.DataFrame({'Title':name, 'Ratings':rating, "Year_of_Release":year})
```

```python
imdb_50
```

### Q3. Write a python program to display IMDB’s Top rated 50 Indian movies’ data (i.e. name, rating, year of release) and make data frame.

```python
url = "https://www.imdb.com/india/top-rated-indian-movies/"
page = requests.get(url)
soup = BeautifulSoup(page.content)

# creating empty lists
title = []
year = []
rating = []

for i in soup.find_all('td', class_='titleColumn'):
    title.append(i.text.split('\n')[2].strip())
    
for y in soup.find_all('span', class_='secondaryInfo'):
    year.append(y.text)
    
for rate in soup.find_all('td', class_='ratingColumn imdbRating'):
    rating.append(rate.text.strip())
```

```python
india_imdb = pd.DataFrame({'Title':title, 'Year_of_Release':year, 'Ratings':rating})
india_50 = india_imdb.head(50)
india_50
```

### Q4. Write a python program to display list of respected former presidents of India(i.e. Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
url = "https://presidentofindia.nic.in/former-presidents.htm"
page = requests.get(url)
soup = BeautifulSoup(page.content)
```

```python
# creating empty lists
name = []
term =[]

for i in soup.find_all('div', class_="presidentListing"):
    name.append(i.text.split('\n')[1].split('(')[0].strip())
    term.append(i.text.split('\n')[2].replace('Term of Office: ', ""))
    
```

```python
President_term = pd.DataFrame({'Name':name,'Term_of_Office': term})
```

### Q5. Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make dataframe.

### Part 1) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
url = 'https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
page = requests.get(url)
soup = BeautifulSoup(page.content)

# creating empty lists
team = []
matches = []
points = []
rating = []
for i in soup.find_all('tr', class_="table-body"):
    team.append(i.text.strip().split('\n')[3])
for m in soup.find_all('tr', class_='table-body'):
    matches.append(m.text.strip().split('\n')[6])
for p in soup.find_all('tr', class_="table-body"):
    points.append(p.text.strip().split('\n')[7])
for rate in soup.find_all('tr', class_="table-body"):
    rating.append(rate.text.strip().split('\n')[8])
    
odi_men = pd.DataFrame({'Team':team,'Matches':matches,'Points':points,"Ratings":rating})
```

```python
# value in the banner

match = soup.find('tr', class_='rankings-block__banner')
match = match.text.strip().split('\n')
banner = [match[3],match[6],match[7],match[9].strip()]

banner_df = pd.DataFrame([banner], columns=['Team','Matches','Points','Ratings'])
```

```python
# Creating new dataframe by concatenating both datframes

odi_men = pd.concat([banner_df,odi_men], ignore_index = True)

odi_men_10 = odi_men.head(10)
odi_men_10
```

### Q7. Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame- i)Headline ii) Time iii) News Link.

```python
url = "https://www.cnbc.com/world/?region=world"
page = requests.get(url)
soup = BeautifulSoup(page.content)

# creating empty lists and adding values to it
headline = []
time = []
links =[]

for i in soup.find_all('a', class_="LatestNews-headline"):
    headline.append(i.text)
for i in soup.find_all('time', class_="LatestNews-timestamp"):
    time.append(i.text)
for link in soup.find_all('a', class_="LatestNews-headline"):
    links.append(link.get('href'))
    
# creating dataframe

news = pd.DataFrame({'Headline':headline, 'Time_stamp':time, 'Link':links})
news
```

### Q8. Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles. Scrape below mentioned details and make data frame- i)Paper Title ii) Authors iii) Published Date iv) Paper URL

```python
url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"
page = requests.get(url)
soup = BeautifulSoup(page.content)

# creating empty lists
title = []
author = []
date =[]
links = []

for i in soup.find_all('h2',{'class':'sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg'}):
    title.append(i.text.strip())
    
for j in soup.find_all('span',{'class':'sc-1w3fpd7-0 dnCnAO'}):
    author.append(j.text)

for d in soup.find_all('span',{'class':'sc-1thf9ly-2 dvggWt'}):
    date.append(d.text)
    
for link in soup.find_all('a', href = True, class_ = 'sc-5smygv-0 fIXTHm'):
    links.append(link.get('href'))

    
# creating dataframe

most_dwnld = pd.DataFrame({'Paper_Title':title,'Authors':author,'Published_Date':date,'Paper_URL':links})
most_dwnld
```

### Q9. Write a python program to scrape mentioned details from dineout.co.in and make data frame i)Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL.

```python
url = 'https://www.dineout.co.in/delhi-restaurants/buffet-special'
page = requests.get(url)
soup = BeautifulSoup(page.content)

# creating empty lists
name = []
cuisine = []
location = []
rating = []
image = []

for n in soup.find_all('a', class_='restnt-name ellipsis'):
    name.append(n.text.strip())
    
for c in soup.find_all('span',class_='double-line-ellipsis'):
    cuisine.append(c.text.split('|')[1].strip())
    
for l in soup.find_all('div', class_='restnt-loc ellipsis'):
    location.append(l.text.strip())
    
for r in soup.find_all('div',class_='restnt-rating rating-4'):
    rating.append(r.text.strip())
    
for img in soup.find_all('img', class_='no-img'):
    image.append(img.get('data-src'))
    
    
# creating dataframe

dine_out = pd.DataFrame({"Restraunt_Name":name, "Cuisine":cuisine, "Location":location, "Ratings":rating, "Image_URL":image})
dine_out
```
